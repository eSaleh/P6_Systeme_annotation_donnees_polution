On lit et observe les données pm2_5 en fonction du temps

```python
import pandas as pd
import numpy as np

def to_df(json):
    list_all=[]
    for column in json.keys():
        list_all.append(json[column])
    return pd.DataFrame(np.array(list_all).T,columns=json.keys())

data_341_20240326_realtime=to_df(data_341_20240326_realtime["data"])
data_340_20240326_realtime=to_df(data_340_20240326_realtime["data"])
data_339_20240326_realtime=to_df(data_339_20240326_realtime["data"])
data_338_20240326_realtime=to_df(data_338_20240326_realtime["data"])

data_341_20240326=to_df(data_341_20240326["data"])
data_340_20240326=to_df(data_340_20240326["data"])
data_339_20240326=to_df(data_339_20240326["data"])
data_338_20240326=to_df(data_338_20240326["data"])
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=data_341_20240326 ['time'], y=data_341_20240326 ['pm25'], mode='lines',name="341"))
fig.add_trace(go.Scatter(x=data_340_20240326 ['time'], y=data_340_20240326 ['pm25'], mode='lines',name="340"))
fig.add_trace(go.Scatter(x=data_339_20240326 ['time'], y=data_339_20240326 ['pm25'], mode='lines',name="339"))
fig.add_trace(go.Scatter(x=data_338_20240326 ['time'], y=data_338_20240326 ['pm25'], mode='lines',name="338"))

fig.update_layout(legend_title_text = "Pm2_5 vs time ")
fig.update_xaxes(title_text="Time (realtime=false, échantillonage 10min)")
fig.update_yaxes(title_text="Pm2_5")
fig.show()
```
